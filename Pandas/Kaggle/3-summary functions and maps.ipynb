```python
import pandas as pd

data = {
    "name": [
        "Alice", "Bob", "Charlie", "David", "Eve",
        "Frank", "Grace", "Helen", "Ian", "Jack",
        "Kate", "Liam", "Mona", "Nina", "Oscar",
        "Paul", "Quinn", "Rose", "Steve", "Tina"
    ],
    "family": [
        "Smith", "Johnson", "Williams", "Brown", "Jones",
        "Garcia", "Martinez", "Hernandez", "Lopez", "Wilson",
        "Anderson", "Thomas", "Taylor", "Moore", "Jackson",
        "Martin", "Lee", "Perez", "Clark", "Walker"
    ],
    "age": [25 + i % 10 for i in range(20)],  # Generate ages using a pattern
    "salary": [30000 + (i * 4000) for i in range(20)]  # Generate salaries in increments
}
df = pd.DataFrame(data)
df
```

```python
df["new salary"] = df["salary"].apply(lambda x: x / 1000)
df
```

```python
df["over or under"] = df["new salary"].apply(lambda x: 'under' if x < 45.0 else 'over')
df
```

```python
df['Name_Age'] = df.apply(lambda row: f"{row['name']} - {row['age']}", axis=1)
df
```

```python
def double_salary(salary):
    return salary * 2


df['double salary'] = df['salary'].apply(double_salary)
df
```

7. we'd like to host these wine reviews on our website, but rate system changing from 80 to 100 points is too hard to understand - we'd like to
translate them to simpler star ratings. A score of 95 and higher counts as 3 stars, a score of at least 85 but less than 95 is 2 stars, any other
score is 1 star. alos Canadian Vintners Association bought alot of ads on the site, so any wines from Canada should automatically get 3 stars,
regardless of points. create a Series "star_ratings" with the number of stars corresponding to each review in the dataset

```python
def star_giver(rows):
    if rows['country'] == 'Canada':
        return 3
    elif rows['points'] >= 95:
        return 3
    elif 85 <= rows['points'] < 95:
        return 2
    else:
        return 1


star_ratings = reviews.apply(star_giver, axis=1)

star_ratings.name = "Stars"
star_ratings
```
