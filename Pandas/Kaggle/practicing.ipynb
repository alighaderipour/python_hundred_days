```python
import pandas as pd

data = {
    "name": [
        "Alice", "Bob", "Charlie", "David", "Eve",
        "Frank", "Grace", "Helen", "Ian", "Jack",
        "Kate", "Liam", "Mona", "Nina", "Oscar",
        "Paul", "Quinn", "Rose", "Steve", "Tina"
    ],
    "family": [
        "Smith", "Johnson", "Williams", "Brown", "Jones",
        "Garcia", "Martinez", "Hernandez", "Lopez", "Wilson",
        "Anderson", "Thomas", "Taylor", "Moore", "Jackson",
        "Martin", "Lee", "Perez", "Clark", "Walker"
    ],
    "age": [25 + i % 10 for i in range(20)],  # Generate ages using a pattern
    "salary": [30000 + (i * 4000) for i in range(20)]  # Generate salaries in increments
}

df = pd.DataFrame(data)
df
```

```python
df["new salary"] = df["salary"].apply(lambda x: x / 1000)
df
```

```python
df["over or under"] = df["salary"].apply(lambda x: 'under' if x < 30 else 'over')
```

```python
df['Name_Age'] = df.apply(lambda row: f"{row['name']} - {row['age']}", axis=1)
df
```
